Step 1: Generate 1000 random numbers over the range 0-100

Use random to generate 1000 random number of list, and store in x

```python
import matplotlib.pyplot as plt
import json
import random

# generate 1000 random points x
x = [random.uniform(0,100) for i in range(1000)]

# save list of points
with open('random_points_x.json', 'w') as f:
    f.write(json.dumps(x))
    
# visualization
plt.scatter(range(1000),x,s=0.3)
plt.title('random_point_x_Visualization')
plt.show()
```

Step 2: Generate new numbers from the original 1000 using the equation y=3x+6

Import x from a.py, get what we generated in a.py, and generate the corresponding y list(also 1000 numbers).

```python
# run b.py to generate 1000 random points y
y = [3 * i + 6 for i in x]

# save list of points
with open('random_points_y.json', 'w') as f:
    f.write(json.dumps(y))
    
# visualization
plt.scatter(range(1000),y,s=0.3)
plt.title('random_point_y_Visualization')
plt.show()
```
